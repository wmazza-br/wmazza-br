Tratando alguns dados estatisticamente

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (14,6))
```

```python
dados.boxplot(['obitos_novos'])
```

```python
dados[dados['obitos_novos'] >= 200].head()
```

```python
valor = dados['obitos_novos']
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
selecao = (valor >=limite_inferior) & (valor <= limite_superior)
dados_new = dados[selecao]
```

```python
dados.hist(['obitos_novos']) #histograma dos dados originais
dados_new.hist(['obitos_novos']) #histograma dos dados sem outliers
```

```python
dados.boxplot(['obitos_novos'], by = ['nome_munic'])
```

```python
grupo_tipo = dados.groupby('nome_munic')['obitos_novos']
```

```python
Q1 =  grupo_tipo.quantile(.25)
Q3 =  grupo_tipo.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
limite_superior['São Paulo']
```

```python
for tipo in grupo_tipo.groups.keys():
    print(tipo)
```

```python
dados_new = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
    eh_tipo = dados['nome_munic'] == tipo
    eh_dentro_limite = (dados['obitos_novos'] >= limite_inferior[tipo]) & (dados['obitos_novos'] <= limite_superior[tipo])
    selecao = eh_tipo & eh_dentro_limite
    dados_selecao = dados[selecao]
    dados_new = pd.concat([dados_new, dados_selecao])
```

```python
dados_new.boxplot(['obitos_novos'], by = ['nome_munic'])
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15,8))
```

```python
dados = pd.read_csv("20211219_dados_covid_municipios_sp.csv", encoding= "ISO-8859-1", sep = ';')

dados.head()
```

```python
area = plt.figure() 
```

```python
g1 = area.add_subplot(2, 2, 1)
g2 = area.add_subplot(2, 2, 2)
g3 = area.add_subplot(2, 2, 3)
g4 = area.add_subplot(2, 2, 4)
```

```python
g1.scatter(dados['casos_novos'], dados['obitos_novos'])
g1.set_title('Casos X Óbitos')

g2.hist(dados['obitos_novos'])
g2.set_title('Histograma')

dados_g3 = dados['obitos_novos'].sample(100) 
dados_g3.index = range(dados_g3.shape[0])
g3.plot(dados_g3)
g3.set_title('Amostra (Valor)')

grupo = dados.groupby('nome_ra')['obitos_novos']
label = grupo.mean().index
valores = grupo.mean().values
g4.bar(label, valores)
g4.set_title('Óbitos por regionais')
```
